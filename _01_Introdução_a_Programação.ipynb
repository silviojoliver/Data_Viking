**Variáveis**
<br><br>
Até agora, você usou código para fazer um cálculo e imprimir o resultado, e o resultado não foi salvo em nenhum lugar. No entanto, você pode imaginar que deseja salvar o resultado para trabalhar com ele mais tarde. Para isso, você precisará usar variáveis.

Criando variáveis
A próxima célula de código cria uma variável chamada test_var e atribui a ela o valor que obtemos quando adicionamos 5 a 4.

Em seguida, imprimimos o valor atribuído à variável, que é 9.

```python
Nome = 'Odemir'
Endereco = 'Brasil'
Valor = 10
Equacao = 2 + 4 * 5
Resultado = Valor * Equacao

print( Nome, Endereco, Valor, Equacao, Resultado )
```

Em geral, para trabalhar com uma variável, você precisa começar selecionando o nome que deseja usar. Nomes de variáveis são idealmente curtos e descritivos. Eles também precisam atender a vários requisitos:

Eles não podem ter espaços (por exemplo, test var não é permitido)
Eles podem incluir apenas letras, números e sublinhados (por exemplo, test_var! não é permitido)
Eles devem começar com uma letra ou sublinhado (por exemplo, 1_var não é permitido)
Então, para criar a variável, você precisa usar = para atribuir o valor que deseja que ela tenha.

Você sempre pode dar uma olhada no valor atribuído à variável usando print() e colocando o nome da variável entre parênteses.

Com o tempo, você aprenderá como selecionar bons nomes para variáveis Python. Não há problema em se sentir desconfortável agora, e a melhor maneira de aprender é apenas visualizando muito código Python!

**Usando várias variáveis**
<br><br>
É comum que o código use várias variáveis. Isso é especialmente útil quando temos que fazer um cálculo longo com várias entradas.

Na próxima célula de código, calculamos o número de segundos em quatro anos. Este cálculo usa cinco entradas.

```python
A = 1
B = 2
C = 3
D = A - B * C

Taxa_Juros = 50
Senha_Acesso = 123456


print( D )
```

```python

N_Anos = 4
Dias_Ano = 365.25
Horas_Dia = 24
Min_Hora = 60
Segundos_Min = 60

Total = Segundos_Min * Min_Hora * Horas_Dia * Dias_Ano * N_Anos
print( Total )
```

Conforme calculado acima, há 126144000 segundos em quatro anos.

Observe que é possível fazer esse cálculo sem variáveis como apenas 60 * 60 * 24 * 365 * 4, mas é muito mais difícil verificar se o cálculo sem variáveis não possui algum erro, pois não é tão legível. Quando usamos variáveis (como num_years, days_per_year etc.), podemos acompanhar melhor cada parte do cálculo e verificar e corrigir erros com mais facilidade.

Observe que é particularmente útil usar variáveis quando os valores das entradas podem mudar. Por exemplo, digamos que queremos melhorar ligeiramente nossa estimativa atualizando o valor do número de dias em um ano de 365 para 365,25, para contabilizar os anos bissextos. Em seguida, podemos alterar o valor atribuído a days_per_year sem alterar nenhuma das outras variáveis e refazer o cálculo.

```python
(10, 1 + 1)
10.89909877
```

Isso se deve ao fato de que no segundo cálculo usamos um número com parte fracionária (365,25), enquanto o primeiro cálculo multiplicou apenas números sem parte fracionária.
